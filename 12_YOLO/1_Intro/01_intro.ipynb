## **Diferencias entre los formatos de cajas:**

xywh, xywhn , xyxy , xyxyn

- **xywh**: Coordenadas absolutas `[x_centro, y_centro, ancho, alto]` en píxeles respecto a la imagen original.
- **xywhn**: Coordenadas normalizadas `[x_centro, y_centro, ancho, alto]` en valores entre 0 y 1, respecto al tamaño de la imagen.
- **xyxy**: Coordenadas absolutas `[x_min, y_min, x_max, y_max]` en píxeles, es decir, esquina superior izquierda y esquina inferior derecha.
- **xyxyn**: Coordenadas normalizadas `[x_min, y_min, x_max, y_max]` en valores entre 0 y 1, respecto al tamaño de la imagen.

Estos formatos permiten trabajar con las cajas de detección en diferentes contextos: píxeles reales o valores normalizados para facilitar el procesamiento y visualización.

# Video

```python
from ultralytics import YOLO
import cv2
import time
```

```python
cap = cv2.VideoCapture("Videos/people.mp4")


while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    cv2.imshow("frame", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    

cap.release()
cv2.destroyAllWindows()
for _ in range(5):
    cv2.waitKey(1)
time.sleep(1)
```

```python
cap = cv2.VideoCapture("Videos/people.mp4")
model = YOLO("yolo11n.pt")

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break
    
    results = model(frame)
    annotated_frame = results[0].plot()
    
    cv2.imshow("frame", frame)
    cv2.imshow("frame", annotated_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
for _ in range(5):
    cv2.waitKey(1)
time.sleep(1)
```
